# 1. 아이 동반 쇼핑시설 데이터 전처리

```python
import pandas as pd
import chardet

# 데이터 불러오기

# 파일의 인코딩 감지
with open('/content/drive/MyDrive/문화 공모전/아이랑 쇼핑시설 목록(2023)/KC_INFANT_SHOPNG_FCLTY_LIST_2023.csv', 'rb') as f:
    result = chardet.detect(f.read())

# 인코딩 감지 결과 출력
print(result)

# 감지된 인코딩으로 파일 읽기
shopping = pd.read_csv('/content/drive/MyDrive/문화 공모전/아이랑 쇼핑시설 목록(2023)/KC_INFANT_SHOPNG_FCLTY_LIST_2023.csv', encoding=result['encoding'])

# 데이터 확인
shopping
```

```python
# 주소 컬럼 첫 번째 행정구역 변환 및 두 번째 행정구역까지 추출하기
def extract_second_level(address):
    # 주소를 공백으로 분할
    new_addr = address.split()

    # 분할된 첫 번째 단어에 따라 지역명을 통합한 이름으로 치환
    regions = {
        '서울': '서울', '서울특별시': '서울',
        '부산': '부산', '부산광역시': '부산',
        '대구': '대구', '대구광역시': '대구',
        '인천': '인천', '인천광역시': '인천',
        '광주': '광주', '광주광역시': '광주',
        '대전': '대전', '대전광역시': '대전',
        '울산': '울산', '울산광역시': '울산',
        '세종': '세종', '세종특별자치시': '세종',
        '경기': '경기', '경기도': '경기',
        '강원': '강원', '강원도': '강원', '강원특별자치도': '강원',
        '충북': '충북', '충청북도': '충북',
        '충남': '충남', '충청남도': '충남',
        '전북': '전북', '전라북도': '전북', '전북특별자치도': '전북',
        '전남': '전남', '전라남도': '전남',
        '경북': '경북', '경상북도': '경북',
        '경남': '경남', '경상남도': '경남',
        '제주': '제주', '제주도': '제주', '제주특별자치도': '제주'
    }

    if new_addr[0] in regions:
        new_addr[0] = regions[new_addr[0]]

    if len(new_addr) >= 2:
        return new_addr[0] + " " + new_addr[1]
    else:
        return new_addr[0]
```

```python
# 시설명, 주소 컬럼만 남기기
shopping_filtered = shopping[['FCLTY_NM', 'LNM_ADDR']]

# 주소 컬럼 두 번째 행정구역까지 자르기
shopping_filtered['LNM_ADDR'] = shopping_filtered['LNM_ADDR'].apply(extract_second_level)

shopping_filtered
```

```python
# 지역별 시설 개수 세기
region_counts = shopping_filtered['LNM_ADDR'].value_counts()

region_counts
```

```python
# 데이터 프레임으로 변환하기
region_counts_df = pd.DataFrame({'지역': region_counts.index, '개수': region_counts.values})

# 시설 개수 기준으로 내림차순 정렬하기
region_counts_df = region_counts_df.sort_values(by='개수', ascending=False)

# csv 파일로 저장하기
region_counts_df.to_csv('아이 동반 쇼핑시설 지역별 시설 개수.csv', index=False)

region_counts_df
```

# 2. 전국 주차장 데이터 전처리


```python
import pandas as pd
import chardet

# 데이터 불러오기

# 파일의 인코딩 감지
with open('/content/drive/MyDrive/문화 공모전/전국주차장정보표준데이터/전국주차장정보표준데이터.csv', 'rb') as f:
    result = chardet.detect(f.read())

# 인코딩 감지 결과 출력
print(result)

# 감지된 인코딩으로 파일 읽기
parking = pd.read_csv('/content/drive/MyDrive/문화 공모전/전국주차장정보표준데이터/전국주차장정보표준데이터.csv', encoding=result['encoding'])

# 데이터 확인
parking
```

```python
# 시설명, 주소 컬럼만 남기기
parking_filtered = parking[['주차장명', '소재지도로명주소', '소재지지번주소']]

# NaN 값을 빈 문자열로 대체
parking_filtered['소재지지번주소'] = parking_filtered['소재지지번주소'].fillna('')

# 주소 컬럼 생성하기
parking_filtered['소재지주소'] = parking_filtered.apply(
    lambda row: row['소재지도로명주소'] if row['소재지지번주소'] == '' else row['소재지지번주소'], axis=1
)

# 주소 컬럼 두 번째 행정구역까지 자르기
parking_filtered['소재지주소'] = parking_filtered['소재지주소'].apply(extract_second_level)

parking_filtered
```

```python
# 지역별 시설 개수 세기
region_counts = parking_filtered['소재지주소'].value_counts()

region_counts
```

```python
# 데이터 프레임으로 변환하기
region_counts_df = pd.DataFrame({'지역': region_counts.index, '개수': region_counts.values})

# 시설 개수 기준으로 내림차순 정렬하기
region_counts_df = region_counts_df.sort_values(by='개수', ascending=False)

# csv 파일로 저장하기
region_counts_df.to_csv('주차장 지역별 개수.csv', index=False)

region_counts_df
```
